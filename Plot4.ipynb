```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Read the data
df = pd.read_csv('household_power_consumption.txt', sep=";",
                   na_values=["?"],
                   parse_dates={'DateTime': ['Date', 'Time']},
                   keep_date_col=True,
                   dayfirst=True
                )
```

```python
# Filter data for dates 2007-02-01 and 2007-02-02

df = df[(df['DateTime'] >= '2007-02-01') & (df['DateTime'] <= '2007-02-02')]

df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot 1: Global Active Power
axs[0, 0].plot(df['DateTime'], df['Global_active_power'], color='black')
axs[0, 0].set_title('Global Active Power')
axs[0, 0].set_xlabel('DateTime')
axs[0, 0].set_ylabel('Global Active Power (Kilowatts)')

# Plot 2: Energy Sub Metering
axs[0, 1].plot(df['DateTime'], df['Sub_metering_1'], color='black', label='Sub_metering_1')
axs[0, 1].plot(df['DateTime'], df['Sub_metering_2'], color='red', label='Sub_metering_2')
axs[0, 1].plot(df['DateTime'], df['Sub_metering_3'], color='blue', label='Sub_metering_3')
axs[0, 1].set_title('Energy Sub Metering')
axs[0, 1].set_xlabel('DateTime')
axs[0, 1].set_ylabel('Energy Sub Metering')
axs[0, 1].legend()

# Plot 3: Voltage
axs[1, 0].plot(df['DateTime'], df['Voltage'], color='black')
axs[1, 0].set_title('Voltage')
axs[1, 0].set_xlabel('DateTime')
axs[1, 0].set_ylabel('Voltage')

# Plot 4: Global Reactive Power
axs[1, 1].plot(df['DateTime'], df['Global_reactive_power'], color='black')
axs[1, 1].set_title('Global Reactive Power')
axs[1, 1].set_xlabel('DateTime')
axs[1, 1].set_ylabel('Global Reactive Power')

# Adjust layout
plt.tight_layout()

# Save plot to PNG
plt.savefig('plot4.png', dpi=100)
```
