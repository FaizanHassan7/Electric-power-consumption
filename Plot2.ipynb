```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Read the data
df = pd.read_csv('household_power_consumption.txt', sep=";",
                   na_values=["?"],
                   parse_dates={'DateTime': ['Date', 'Time']},
                   keep_date_col=True,
                   dayfirst=True
                )
```

```python
# Filter data for dates 2007-02-01 and 2007-02-02

df = df[(df['DateTime'] >= '2007-02-01') & (df['DateTime'] <= '2007-02-02')]

df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
# Plot 2
plt.figure(figsize=(8, 6))
plt.plot(df['DateTime'], df['Global_active_power'], color='black', linestyle='-')  # 'l' style is equivalent to '-' in Matplotlib
plt.title('Global Active Power')
plt.xlabel('DateTime')
plt.ylabel('Global Active Power (kilowatts)')
plt.tight_layout()
# Save plot to PNG
plt.savefig('plot2.png', dpi=100)
```
