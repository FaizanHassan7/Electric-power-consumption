```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Read the data
df = pd.read_csv('household_power_consumption.txt', sep=";",
                   na_values=["?"],
                   parse_dates={'DateTime': ['Date', 'Time']},
                   keep_date_col=True,
                   dayfirst=True
                )
```

```python
# Filter data for dates 2007-02-01 and 2007-02-02

df = df[(df['DateTime'] >= '2007-02-01') & (df['DateTime'] <= '2007-02-02')]

df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
# Plot 3
plt.figure(figsize=(8, 6))
plt.plot(df['DateTime'], df['Sub_metering_1'], color='black', label='Sub_metering_1')
plt.plot(df['DateTime'], df['Sub_metering_2'], color='red', label='Sub_metering_2')
plt.plot(df['DateTime'], df['Sub_metering_3'], color='blue', label='Sub_metering_3')
plt.title('Energy Sub Metering')
plt.xlabel('Date')
plt.ylabel('Energy sub metering')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()

# Save plot to PNG
plt.savefig('plot3.png', dpi=100)
```
